```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

**Use below dataframe for plots**

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

**Bar plots**

# 1. Set darkgrid style from seaborn

```python
sns.set(style='darkgrid')
```

# 2. Make a bar plot

```python
plt.figure(figsize=(8,6), dpi=500)
fig, ax = plt.subplots()
ax.bar(x=df.index, height=df.income, width=0.3, align='edge', label='income')
ax.bar(x=df.index, height=df.expense, width=-0.3, align='edge', label='expense')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# 3. Stack the bars

```python
plt.figure(figsize=(8,6), dpi=500)
fig, ax = plt.subplots()
ax.bar(x=df.index, height=df.income, width=0.4, align='center', label='income')
ax.bar(x=df.index, height=df.expense, width=0.4, align='center', label='expense', bottom=df.income)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# 4. Rotate the labels and set figsize

```python
sns.set(rc={'figure.figsize':(10,8)})
fig, ax = plt.subplots()
ax.bar(x=df.index, height=df.income, width=0.5, align='center', label='income')
ax.bar(x=df.index, height=df.expense, width=0.5, align='center', label='expense', bottom=df.income)
plt.legend()
plt.show()
```

# 5. Replace the labels by "Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"

```python
sns.set(rc={'figure.figsize':(8,6)})
fig, ax = plt.subplots()
ax.bar(x=df.index, height=df.income, width=0.5, align='center', label='income')
ax.bar(x=df.index, height=df.expense, width=0.5, align='center', label='expense', bottom=df.income)
ax.set_xticklabels(['Game-1', 'Game-2', 'Game-3', 'Game-4', 'Game-5', 'Game-6', 'Game-7'])
plt.legend()
plt.show()
```

# 6. Unstuck the bars, annotate the hight of the bars on top of them

```python
sns.set(rc={'figure.figsize':(8,6)})
fig, ax = plt.subplots()

for bar in ax.bar(x=df.index, height=df.income, width=0.3, align='edge', label='income').patches:
    ax.annotate(bar.get_height(),(bar.get_x() , bar.get_height() + 1))
for bar in ax.bar(x=df.index, height=df.expense, width=-0.3, align='edge', label='expense').patches:
    ax.annotate(bar.get_height(),(bar.get_x() - 0.25 , bar.get_height() + 1))
ax.set_xticklabels(['Game-1', 'Game-2', 'Game-3', 'Game-4', 'Game-5', 'Game-6', 'Game-7'])
plt.legend()
plt.show()
```

# 12. Seperate the the boxplot above using size columns

```python
sns.set(style='whitegrid')
fig,ax = plt.subplots()

size1 = tips.loc[tips['size'] == 1,'total_bill']
size2 = tips.loc[tips['size'] == 2,'total_bill']
size3 = tips.loc[tips['size'] == 3,'total_bill']
size4 = tips.loc[tips['size'] == 4,'total_bill']
size5 = tips.loc[tips['size'] == 5,'total_bill']
size6 = tips.loc[tips['size'] == 6,'total_bill']
data=[size1,size2,size3,size4,size5,size6]
ax.boxplot(data, vert=True)

ax.set_xlabel('size')
fig.suptitle('Boxplot group by size')
plt.show()
```

# 13. Make the same plot using seaborn

```python
sns.boxplot(data=data)
plt.xlabel('size')
plt.ylabel('total_bill');
```
